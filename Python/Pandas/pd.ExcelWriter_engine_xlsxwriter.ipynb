```python
from xlsxwriter.format import Format
from xlsxwriter.workbook import Workbook
from xlsxwriter.worksheet import Worksheet


def _writer(
    df: pd.DataFrame,
    writer: pd.ExcelWriter,
    sheet_name: str,
    start_row: int = 0,
    start_col: int = 0,
    header: bool = True,
    header_length: int = 1,
) -> None:
    df.to_excel(writer, sheet_name=sheet_name, startrow=start_row, startcol=start_col, header=header, index=False)

    workbook: Workbook = writer.book
    worksheet: Worksheet = writer.sheets[sheet_name]

    max_row, max_col = df.shape[0] + start_row, df.shape[1] + start_col

    start_header = start_row + header_length
    start_value = start_header + 1

    # Write White
    white_format: Format = workbook.add_format()
    white_format.set_bg_color('#FFFFFF')

    for row_num in range(0, start_header):
        for col_num in range(0, max_col):
            worksheet.write_blank(row_num, col_num, None, white_format)

        if start_col > 0:
            worksheet.write_blank(row_num, 0, None, white_format)

    # Write header
    if header:
        header_format = workbook.add_format()
        header_format.set_font_name('Arial')

        for col_num, value in enumerate(df.columns.values):
            worksheet.write(start_row, start_col + col_num, value, header_format)

            # Merge cells if header_length > 1
            if header_length > 1:
                worksheet.merge_range(
                    start_row,
                    start_col + col_num,
                    start_row + header_length - 1,
                    start_col + col_num,
                    value,
                    header_format,
                )

        # Set AutoFilter
        worksheet.autofilter(start_row, start_col, start_row, max_col - 1)

    cell_format = workbook.add_format()
    cell_format.set_font_name('Arial')

    # Font settings
    # cell_format.set_font_name('Arial')
    # cell_format.set_font_size(12)
    # cell_format.set_font_color('black')
    # cell_format.set_bold()
    # cell_format.set_italic()
    # cell_format.set_underline()
    # cell_format.set_font_strikeout()
    # cell_format.set_font_script(1)  # 1 for superscript, 2 for subscript

    # Number format
    # cell_format.set_num_format('0.00')

    # Protection settings
    # cell_format.set_locked()
    # cell_format.set_hidden()

    # Alignment settings
    # cell_format.set_align('center')
    # cell_format.set_valign('vcenter')
    # cell_format.set_rotation(90)
    # cell_format.set_text_wrap()
    # cell_format.set_reading_order(2)  # 2 for right-to-left
    # cell_format.set_text_justlast()
    # cell_format.set_center_across()
    # cell_format.set_indent(1)
    # cell_format.set_shrink()

    # Pattern settings
    # cell_format.set_pattern(1)  # 1 for solid fill
    # cell_format.set_bg_color('#FFFFFF')
    # cell_format.set_fg_color('#000000')

    # Border settings
    # cell_format.set_border(1)
    # cell_format.set_bottom(1)
    # cell_format.set_left(1)
    # cell_format.set_right(1)
    # cell_format.set_border_color('#000000')
    # cell_format.set_bottom_color('#000000')
    # cell_format.set_top_color('#000000')
    # cell_format.set_left_color('#000000')
    # cell_format.set_right_color('#000000')

    return
```
